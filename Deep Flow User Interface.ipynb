## Deep Learning

You may create a deep neural network either by defining your own architecture, or setting `param='auto'` to let DeepFlow automatically create the best network possible using hyperparameter optimization.

In this prototype demonstration, we will use the following X -> Y mapping. Notice how easy it is to modify the pandas table style dataset object.

X = (open_price, previous_day_open_price, previous_day_high_price, previous_day_low_price, previous_day_close_price, previous_day_volume)

Y = (close_price)

Based on the shape of X, the created deep network will have 6 subunits. We can choose the number of layers

We form a train, test, and validation set. Here, we choose to split the dataset into 56% train, 33% test, and 10% validation.

```python
apple_pre = apple[apple['quotedate']<'2009-12-31']
apple_post = apple[apple['quotedate']>'2009-12-31']

def prepare_apple_dataset(apple):
    apple = apple[['open_price', 
                         'high_price', 
                         'low_price', 
                         'close_price', 
                         'volume']]

    apple['prev_open'] = apple.open_price.shift(1)
    apple['prev_high'] = apple.high_price.shift(1)
    apple['prev_low'] = apple.low_price.shift(1)
    apple['prev_close'] = apple.close_price.shift(1)
    apple['prev_volume'] = apple.volume.shift(1)

    apple = apple[['open_price',
                              'prev_open',
                              'prev_high', 
                              'prev_low', 
                              'prev_close', 
                              'prev_volume',
                              'close_price']]

    return dataset.remove_missing_values(apple)

clean_apple_pre = prepare_apple_dataset(apple_pre)
clean_apple_post = prepare_apple_dataset(apple_post)

train, test, validation = dataset.split_train_test_validation(
                                            clean_apple_pre, 
                                            test_size=0.33, 
                                            validation_size=0.15)

viewer.table(clean_apple_pre)
```

### Deep Feedforward Network
#### Create and Train the Deep Feedforward Network

```python
deepflow = df.Learn(dataset)

params = {'units': [6, 7, 6],
          'steps': 1000,
          'batch_size': 100}

trainX, trainY = train[['open_price',
                        'prev_open',
                        'prev_high', 
                        'prev_low', 
                        'prev_close', 
                        'prev_volume']], train[['close_price']]

deepflow.feedforward(mode='train', 
                     param=params, 
                     x=trainX, 
                     y=trainY)
```

#### Evaluate the Deep Feedforward Network

```python
testX, testY = test[['open_price',
                        'prev_open',
                        'prev_high', 
                        'prev_low', 
                        'prev_close', 
                        'prev_volume']], test[['close_price']]

deepflow.feedforward(mode='test', 
                     param=params, 
                     x=testX, 
                     y=testY)
```

#### Compare the Deep Feedforward Network with real life

```python
validationX, validationY = validation[['open_price',
                        'prev_open',
                        'prev_high', 
                        'prev_low', 
                        'prev_close', 
                        'prev_volume']], validation[['close_price']]

predictions = [p for p in deepflow.feedforward(mode='predict',
                     param=params,
                     x=validationX)]

for ypred, yact in zip(predictions, validationY.values.tolist()):
    print('Predicted: {} v.s. Expected: {}'.format(
                    ypred['predictions'], yact))
```

```python
rmse = deepflow.rmse(predictions, validationY)
print('Resultant RMSE: {}'.format(rmse))
```
