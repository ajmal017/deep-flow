## Deep Learning

You may create a deep neural network either by defining your own architecture, or setting `param='auto'` to let DeepFlow automatically create the best network possible using hyperparameter optimization.

In this prototype demonstration, we will use the following X -> Y mapping. Notice how easy it is to modify the pandas table style dataset object.

X = (open_price, previous_day_open_price, previous_day_high_price, previous_day_low_price, previous_day_close_price, previous_day_volume)

Y = (close_price)

Based on the shape of X, the created deep network will have 6 subunits. We can choose the number of layers

We form a train, test, and validation set. Here, we choose to split the dataset into 56% train, 33% test, and 10% validation. This is useful when you're interested in comparing different models.

```python
# split apple stock data into two sets: 
# 1) 2002-2010 (model building)
# 2) 2010-2011 (comparison against s&p500)
apple_pre = apple[apple['quotedate']<'2009-12-31']
apple_post = apple[apple['quotedate']>'2009-12-31']

def prepare_apple_dataset(apple):
    """ This function creates the apple dataset as specified.
        Parameters:
        apple: pandas dataframe object of apple stock dataset
        
        Returns:
        new_apple: processed pandas dataframe object
    """
    # select features
    apple = apple[['quotedate',
                   'open_price', 
                   'high_price', 
                   'low_price', 
                   'close_price', 
                   'volume']]

    # using pandas shift(), create new "last day" features
    apple['prev_open'] = apple.open_price.shift(1)
    apple['prev_high'] = apple.high_price.shift(1)
    apple['prev_low'] = apple.low_price.shift(1)
    apple['prev_close'] = apple.close_price.shift(1)
    apple['prev_volume'] = apple.volume.shift(1)

    # select final feature set
    apple = apple[['quotedate',
                   'open_price',
                   'prev_open',
                   'prev_high', 
                   'prev_low', 
                   'prev_close', 
                   'prev_volume',
                   'close_price']]

    # data preprocessing
    return dataset.remove_missing_values(apple)

clean_apple_pre = prepare_apple_dataset(apple_pre)
clean_apple_post = prepare_apple_dataset(apple_post)

# set test and validation proportions by key word argument
train, test, validation = dataset.split_train_test_validation(
                                            clean_apple_pre, 
                                            test_size=0.33, 
                                            validation_size=0.15)

viewer.table(clean_apple_pre)
```

### Deep Feedforward Network
#### Create and Train the Deep Feedforward Network

```python
# initialize deepflow.learn object
deepflow = df.Learn(dataset)

# create parameters dictionary
params = {'units': [6, 5, 6],
          'steps': 10000,
          'batch_size': 1000}

# indicate training set into features and labels
trainX, trainY = train[['open_price',
                        'prev_open',
                        'prev_high', 
                        'prev_low', 
                        'prev_close', 
                        'prev_volume']], train[['close_price']]

# train a feedforward network
deepflow.feedforward(mode='train', 
                     param=params, 
                     x=trainX, 
                     y=trainY)
```

#### Evaluate the Deep Feedforward Network

```python
testX, testY = test[['open_price',
                     'prev_open',
                     'prev_high', 
                     'prev_low', 
                     'prev_close', 
                     'prev_volume']], test[['close_price']]

deepflow.feedforward(mode='test', 
                     param=params, 
                     x=testX, 
                     y=testY)
```

#### Compare the Deep Feedforward Network with real life

```python
validationX, validationY = validation[['open_price',
                        'prev_open',
                        'prev_high', 
                        'prev_low', 
                        'prev_close', 
                        'prev_volume']], validation[['close_price']]
 
predictions = deepflow.feedforward(mode='predict',
                                  param=params,
                                  x=validationX)

# for ypred, yact in zip(predictions, validationY.values.tolist()):
#     print('Predicted: {} v.s. Expected: {}'.format(
#                     ypred['predictions'], yact))

viewer.plot([predictions, validationY], 
            x=validation['quotedate'],
            title='Comparison between Predicted and Expected Close Price')
```

### Root Mean Square Error

```python
rmse = deepflow.rmse(predictions, validationY)
print('Resultant RMSE: {}'.format(rmse))
```

## High Frequency Trader

Let's compare the performance of the High Frequency Trader to the S&P 500. We built a deep neural network using AAPL stock from before the start of 2010. Now we will run the model through 2011, with different bankrolls, and calculate the average return. This way we may compare ROI between the two different methods. So far, we have done this much above.

Now we model what an expert financial agent would do with this information to get some sort of idea if it would provide any sort of substantial benefit over traditional investments. Keep in mind that I, the author of this notebook, am not a financial expert. Nor is the algorithm that we will code momentarily. Because of this, we don't really have a true measure of the benefit offered by the DeepFlow library until we put this in the hads of some actual industry experts. Nevertheless...

For each stock price, if the agent predicts a higher close price, they will buy and sell the next day it produces a profit. 

Remember, this is what our dataset looks like:

```python
viewer.table(clean_apple_post)
```

```python
# split into input feature labels for prediction
evalX = clean_apple_post[['open_price',
                          'prev_open',
                          'prev_high', 
                          'prev_low', 
                          'prev_close', 
                          'prev_volume']]

pred = deepflow.feedforward(mode='predict',
                                   param=params,
                                   x=evalX)
```

Let's take a peak at the predictive power of our deep neural net:

```python
viewer.plot([pred, clean_apple_post['close_price']], 
            title='Comparison between Predicted and Expected Close Price')
```

```python
import pandas as pd
import numpy as np
master = pd.concat([pred.reset_index(drop=True)+100, 
           apple_post.reset_index(drop=True)], axis=1)
train_data = None
simulate_data = None
bankrolls = [100, 1000, 100000, 10000000, 1000000000]
final_bankroll = []
stock = 'AAPL'
risk = 0.05
# Try different initial bankroll sizes
for bankroll in bankrolls:
    for i in range(len(master)-1):
        if master['predictions'].iloc[i] > master['open_price'].iloc[i]:
            risk_multiplier = risk*((master['predictions'].iloc[i] - master['open_price'].iloc[i])**2)**0.5
            num_stocks = int(bankroll / master['open_price'].iloc[i] * risk_multiplier)
            bankroll += (master['open_price'].iloc[i+1]-master['open_price'].iloc[i])*num_stocks
            #bankroll += (master['high_price'].iloc[i+1]-master['open_price'].iloc[i])*num_stocks
    final_bankroll.append(bankroll)
    
roi = pd.DataFrame(np.array(final_bankroll)/np.array(bankrolls), 
                   columns=['return on investment'],
                  index=bankrolls)
roi.plot(logx=True, title='Return on Investment v.s. bankroll, AAPL', figsize=(10,8)).set(ylabel='ROI %', xlabel='Initial Bankroll');
```
